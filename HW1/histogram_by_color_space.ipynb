```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Image, display
%matplotlib inline
```

## Sample Image Test

```python
img = cv2.imread('sample_cat.png')
color = ('b','g','r')

plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Sample Image\n")

plt.subplot(122)
plt.title("RGB space - Color Histogram\n")
plt.xlabel("Bins")
plt.ylabel("Frequency")

bin_size = 64

for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[bin_size],[0,256])
    plt.plot(histr, color = col)
    plt.xlim([0,bin_size])
    
plt.tight_layout()
plt.show()
```

## 0. Load the Color Chart Image

```python
img = cv2.imread('color_chart.png')
print(img.shape)
color = ('b','g','r')

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Color Chart Image")

plt.show()

bin_size = 64
```

## 1. RGB space
* __R__: __0-255__
* __G__: __0-255__
* __B__: __0-255__

```python
color = ('b','g','r')
labels = ('B', 'G', 'R')

plt.figure()
plt.title("RGB space - Color Histogram\n")
plt.xlabel("Bins")
plt.ylabel("Frequency")

for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[bin_size],[0,256])
    plt.plot(histr, color = col, label=labels[i])
    plt.xlim([-1,bin_size])
    
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

## 2. HSV space
* __H__: 0-360 =>(H/2)=> __0-180__
* __S__: 0-1 =>(S*255)=> __0-255__
* __V__: 0-255 =>(V*255)=> __0-255__

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

color = ('b','g','r')
labels = ('H', 'S', 'V')

plt.figure()
plt.title("HSV space - Color Histogram\n")
plt.xlabel("Bins")
plt.ylabel("Frequency")

for i,col in enumerate(color):
    histr = cv2.calcHist([hsv],[i],None,[bin_size],[0,256])
    plt.plot(histr, color = col, label=labels[i])
    plt.xlim([-1,bin_size])
    
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    
plt.show()
```

## 3. CIE Lab space
* __L__: 0-100 =>(L*255/100)=> __0-255__
* __a__: -127-127 =>(a+128)=> __1-255__
* __b__: -127-127 =>(b+128)=> __1-255__

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2Lab)

color = ('b','g','r')
labels = ('L', 'a', 'b')

plt.figure()
plt.title("CIE Lab space - Color Histogram\n")
plt.xlabel("Bins")
plt.ylabel("Frequency")

for i,col in enumerate(color):
    histr = cv2.calcHist([hsv],[i],None,[bin_size],[0,256])
    plt.plot(histr, color = col, label=labels[i])
    plt.xlim([-1,bin_size])
    
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    
plt.show()
```
