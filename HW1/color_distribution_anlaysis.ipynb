```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Image, display
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
def convertHEX(bgr):
    hexcode = '#' + format(bgr[0], '02x') + format(bgr[1], '02x') + format(bgr[2], '02x')
    return hexcode
```

## 0. Load the Color Chart Image

```python
img = cv2.imread('color_chart_small.png')
print(img.shape)
color = ('b','g','r')

plt.figure(figsize=(3,3))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Color Chart Image")

plt.show()
```

## RGB space
* __R__: __0-255__
* __G__: __0-255__
* __B__: __0-255__

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(121, projection='3d')
ax.set_xlim3d(0,256)
ax.set_ylim3d(256,0)
ax.set_zlim3d(0,256)
ax.set_xlabel('R')
ax.set_ylabel('G')
ax.set_zlabel('B')

# For each set of style and range settings, plot n random points in the box
for row in img:
    for point in row:
        xs = point[2]
        ys = point[1]
        zs = point[0]
        try:
            ax.scatter(xs, ys, zs, s=50, c=convertHEX([xs,ys,zs]), marker='o')
        except:
            print(xs, ys, zs)
            print(convertHEX([xs,ys,zs]))
            
ax = fig.add_subplot(122, projection='3d')
ax.set_xlim3d(0,256)
ax.set_ylim3d(256,0)
ax.set_zlim3d(0,256)
ax.set_xlabel('R')
ax.set_ylabel('G')
ax.set_zlabel('B')

# For each set of style and range settings, plot n random points in the box
for row in img:
    for point in row:
        xs = point[2]
        ys = point[1]
        zs = point[0]
        try:
            ax.scatter(xs, ys, zs, s=50, c=convertHEX([xs,ys,zs]), marker='o')
        except:
            print(xs, ys, zs)
            print(convertHEX([xs,ys,zs]))
ax.azim = 100
ax.elev = 45

plt.tight_layout()
plt.show()
```
