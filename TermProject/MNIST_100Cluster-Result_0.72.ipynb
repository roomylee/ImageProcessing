# Project. MNIST

```python
import cv2
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import svm
from matplotlib import pyplot as plt
%matplotlib inline
```

# MNIST Date Read

```python
df = pd.read_csv("data/train.csv")
```

```python
label = df['label']
df = df.drop(['label'], axis=1)
df
```

# Make Images

```python
images = []
for i in range(len(df)):
    img = []
    for j in range(0, 28):
        img.append(df.loc[i].tolist()[j*28:(j+1)*28])
    images.append(img)
```

# SIFT & Collect Descriptors

```python
sift = cv2.xfeatures2d.SIFT_create()
descriptor = []
for img in images:
    img = np.array(img, dtype=np.uint8)
    kp, des = sift.detectAndCompute(img,None)
    if des != None:
        for feature in des:
            descriptor.append(feature)
```

```python
len(descriptor)
```

```python
descriptor[0].shape
```

# K Means Clustering for Descriptors

```python
clusters_count = 100
km = KMeans(n_clusters=clusters_count)
km.fit(descriptor)
```

# Classification for Distribution of Descriptors Using SVM

## 1. Train Data

```python
train_X = []
train_y = np.array(label).tolist()
for img in images:
    img = np.array(img, dtype=np.uint8)
    kp, des = sift.detectAndCompute(img,None)
    Dist = [0] * clusters_count
    if des != None:
        result = km.predict(des)
        for cluster in result:
            Dist[cluster] += 1
    train_X.append(Dist)
```

```python
train_X[:5]
```

```python
train_y[:5]
```

## 2. Training Support Vector Machine

```python
clf = svm.LinearSVC()
clf.fit(train_X, train_y)
```

## 3. Test Data

```python
test_df = pd.read_csv("data/test.csv")
```

```python
test_df
```

```python
test_images = []
for i in range(len(test_df)):
    img = []
    for j in range(0, 28):
        img.append(test_df.loc[i].tolist()[j*28:(j+1)*28])
    test_images.append(img)
```

```python
test_X = []
for img in test_images:
    img = np.array(img, dtype=np.uint8)
    kp, des = sift.detectAndCompute(img,None)
    Dist = [0] * clusters_count
    if des != None:
        result = km.predict(des)
        for cluster in result:
            Dist[cluster] += 1
    test_X.append(Dist)
```

## 4. Prediction for Test Data

```python
result = clf.predict(test_X)
```

```python
output = pd.DataFrame(columns=['ImageId', 'Label'])
output['ImageId'] = [i for i in range(1,len(test_df)+1)]
output['Label'] = result.tolist()
```

```python
output
```

## 5. Prediction Result

```python
import matplotlib.gridspec as gridspec
plt.figure(figsize=(10,10))
gs1 = gridspec.GridSpec(3,3)
gs1.update(wspace=0.5, hspace=0.5)
for i in range(9):
    plt.subplot(gs1[i])
    plt.title("Image #%d\nLabel = %d" % (i+1, output['Label'].loc[i]))
    plt.imshow(test_images[i], 'gray')
```
